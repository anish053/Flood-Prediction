```python
import cdsapi
import os
```

```python
cds_url = 'https://cds.climate.copernicus.eu/api'
cds_key = '8adc69dc-4092-4d93-b37a-86b9dfed61db'

```

```python
# Define save paths (inside Google Drive)
base_path = "/content/drive/MyDrive/Flood_Thesis/Data/ERA5_Data"
coarse_path = os.path.join(base_path, "Coarse_ERA5")
fine_path = os.path.join(base_path, "Fine_ERA5")

# Create folders
os.makedirs(coarse_path, exist_ok=True)
os.makedirs(fine_path, exist_ok=True)

# Define variables for download
variables_era5 = {
    "total_precipitation": "tp",
    "2m_temperature": "t2m",
    "volumetric_soil_water_layer_1": "swvl1",
    "10m_u_component_of_wind": "u10",
    "10m_v_component_of_wind": "v10",
    "surface_pressure": "sp",
}

variables_era5_land = {
    "total_precipitation": "tp",
    "2m_temperature": "t2m",
    "volumetric_soil_water_layer_1": "swvl1",
    "10m_u_component_of_wind": "u10",
    "10m_v_component_of_wind": "v10",
    "surface_pressure": "sp",
}

# Kathmandu Valley Bounding Box [North, West, South, East]
area = [27.85, 85.15, 27.4, 85.6]

years = [ "2017", "2018","2019"]
months = [f"{i:02d}" for i in range(1, 13)]
days = [f"{i:02d}" for i in range(1, 32)]
times = ["12:00"]

print("✅ Ready for ERA5 and ERA5-Land download.")
```

```python
import cdsapi

cds = cdsapi.Client(url=cds_url, key=cds_key)

# ERA5 download
for var, short_name in variables_era5.items():
    print(f"📦 Downloading ERA5 {var}...")
    cds.retrieve(
        "reanalysis-era5-single-levels",
        {
            "product_type": "reanalysis",
            "variable": var,
            "year": years,
            "month": months,
            "day": days,
            "time": times,
            "format": "netcdf",
            "area": area,
        },
        f"{coarse_path}/{short_name}_ERA5_2017_2019.nc"
    )

# ERA5-Land download
for var, short_name in variables_era5_land.items():
    print(f"📦 Downloading ERA5-Land {var}...")
    cds.retrieve(
        "reanalysis-era5-land",
        {
            "variable": var,
            "year": years,
            "month": months,
            "day": days,
            "time": times,
            "format": "netcdf",
            "area": area,
        },
        f"{fine_path}/{short_name}_ERA5Land_2017_2019.nc"
    )
```
